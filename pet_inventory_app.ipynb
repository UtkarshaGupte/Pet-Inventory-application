```python
import sys
import re
import datetime
from datetime import date
import pickle
```

```python
class Pet:
    def __init__(self, name, dob, birthWeight, owner):
        try:
            self.__year,self.__month,self.__date = [int(x) for x in dob.split(',')]
            datetime.datetime(self.__year,self.__month,self.__date)
            if len(re.findall(r'\d{4}',str(self.__year))) != 1:
                self.__year=int(datetime.datetime.strptime(str(self.__year),'%y').strftime('%Y'))
        except ValueError:
            print('Invalid Date of Birth')   
            sys.exit()
        else:
            try:
                self.__birthWeight = float(birthWeight)
            except ValueError:
                print('Invalid Birth Weight')
                sys.exit()
            else:
                self.__today_date = date.today()    
                self.__dob = date(self.__year,self.__month,self.__date)
                self.__age = (abs(self.__today_date - self.__dob)).days

                self.__name = name

                self.__owner = owner

        
    def get_name(self):
        return self.__name
    
    def get_dob(self):
        return self.__dob
    
    def get_birthWeight(self):
        return self.__birthWeight
    
    def get_owner(self):
        return self.__owner
    
    def get_age(self):
        return self.__age
    
    def set_name(self, jt):
        self.__name = name
        
    def set_dob(self, sal):
        self.__dob = dob
        
    def set_birthWeight(self, jt):
        self.__birthWeight = birthWeight
        
    def set_owner(self, sal):
        self.__owner = owner
        
    def set_age(self, dob):
        self.__year,self.__month,self.__date = [int(x) for x in dob.split(',')]
        self.__today_date = date.today()    
        self.__dob = date(self.__year,self.__month,self.__date)
        self.__age = (abs(self.__today_date - self.__dob)).days
        
    def get_pet_data(self):
        out_str = 'Pet Name: ' + self.__name + '; Pet DOB: ' + str(self.__dob) + '; Birth Weight: ' + str(self.__birthWeight) + '\n'
        out_str += self.__owner.get_owner_data()
        return out_str
```

```python
class Fish(Pet):
    def __init__(self,name, dob, birthWeight, owner, scaleCondition, length):
        super().__init__(name, dob, birthWeight, owner)
        
        self.__scaleCondition = scaleCondition
        self.__length = length
        
    def get_scaleCondition(self):
        return self.__scaleCondition
    
    def get_length(self):
        return self.__length
    
    
    def set_scaleCondition(self, jt):
        self.__scaleCondition = scaleCondition
        
    def set_length(self, sal):
        self.__length = length
        
    
    def fish_currweight(self):
        age=self.get_age()
        num=90
        while(num<=age and num <=360):
            self._Pet__birthWeight+= 0.04* self._Pet__birthWeight
            num+=90
        return int(self._Pet__birthWeight)
    
    def get_fish_data(self):
        out_str = '\n' + 'Details of this Fish pet are: \n'
        out_str += self.get_pet_data() + '\n'
        out_str+=  'Scale Condition: ' + self.__scaleCondition + '; Length: ' + str(self.__length)
        return out_str
    
    def fish_currweight_str(self):
        return self._Pet__name + ' weighs ' + str(self.fish_currweight()) + ' pounds and is '+ str(self._Pet__age) + ' days old'
        
        
```
